```python
import google.genai as genai
from google.colab import userdata
import random
import numpy as np
import math
import physicist_prompts as prompts

api_key = userdata.get('gemini_token')
client = genai.Client(api_key=api_key)
```

```python
system_prompt = prompts.system_prompt
model_config = genai.types.GenerateContentConfig(
    system_instruction=system_prompt,
)

print(system_prompt)
```

# Run 1

```python
chat = client.chats.create(
    model="gemini-2.5-pro",
    config=model_config
)
```

```python
response = chat.send_message(["We can start now."])
print(response.text)
```

```python
g = 5

mI = 1.0
mG = (math.ceil(mI**0.5))**2

h0 = np.array([5.0,10.0,20.0,40.0,80.0])
a = mG*g/mI
t_ground = np.sqrt((2*h0)/a)

data = np.column_stack((h0, t_ground))
file_name = "run_1_experiment_1.txt"
comment = f'Experiment 1:\nThe time (t) it takes a ball of mass 1 kg to fall to the ground when released from rest from height h.\n\n'
np.savetxt(file_name, data, delimiter=',', header='h (glonks), t (ticks)', comments=comment, fmt='%.3f')
```

```python
experimental_file_1 = client.files.upload(file="/content/run_1_experiment_1.txt")

response = chat.send_message(["Here are the experimental results.", experimental_file_1])
print(response.text)
```
